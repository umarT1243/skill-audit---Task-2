$\huge ii.$ Generate the term ùëõ of a recursive relationship given by


ùëéùëõ = 5ùëéùëõ‚àí1 + 2 , ùëé0 = 2

```python
def recursive(n, a0=2):
    if n == 0:
        return a0
    else:
        return 5 * recursive(n-1) + 2

n = int(input("Enter the value of n: "))
print("The nth term of the recursive relationship is:", recursive(n))
```
